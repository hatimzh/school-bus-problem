```python
import csv
from geopy.geocoders import Nominatim
from geopy.distance import geodesic
import pandas as pd
import random
import folium



# Geolocator to convert addresses to coordinates
geolocator = Nominatim(user_agent="school_bus_route")

students = []
df=pd.read_csv('locations.csv')
students_address=list(zip(df['Address']))
students = list(zip(df['Latitude'], df['Longitude']))

#Dictionary of student's addresses and cordonats
students_dict={cord:'' for cord in students}
for i in students_dict:
    students_dict[i]=geolocator.reverse(f"{i[0]}, {i[1]}").address 
# Define a function to calculate the total distance between a route

def calculate_distance(route):
    total_distance = 0
    for i in range(1, len(route)):
        current_coord = route[i]
        previous_coord = route[i-1]
        distance = geodesic((current_coord[0], current_coord[1]), (previous_coord[0], previous_coord[1])).km
        total_distance += distance
    return total_distance

# Define a function to perform crossover (recombination) between two routes
def crossover(route1, route2):
    n = len(route1)
    crossover_point = random.randint(1, n-1)
    offspring = route1[:crossover_point] + route2[crossover_point:]
    return offspring

# Define a function to perform mutation on a route
def mutate(route):
    mutation_point = random.randint(0, len(route) - 1)
    new_coord = students[0]
    while new_coord in route:
        new_coord =students[0]
    route[mutation_point] = new_coord
    return route
def genetic_algorithm(students, population_size, mutation_rate):
    # Initialize a random population
    population = [list(set(random.sample(students, len(students)))) for i in range(population_size)]

    # Set a stopping criteria
    best_distance = float("inf")
    generations = 0
    while generations < 100:
        new_population = []
        while len(new_population) < population_size:
            # Select parents using roulette wheel selection
            fitness_list = [calculate_distance(x) for x in population]
            fitness_sum = sum(fitness_list)
            normalize_weights = [x/fitness_sum for x in fitness_list]
            parent1 = random.choices(population,weights=normalize_weights)[0]
            parent2 = random.choices(population,weights=normalize_weights)[0]
            # Perform crossover and mutation
            offspring = crossover(parent1, parent2)
            #if random.uniform(0, 1) < mutation_rate:
                #offspring = mutate(offspring)
            new_population.append(offspring)
        population = new_population
        # Get the best route in the current population
        best_route = min(population, key=calculate_distance)
        new_best_distance = calculate_distance(best_route)
        if new_best_distance < best_distance:
            best_distance = new_best_distance
        generations += 1
        print(f'Generation: {generations}  Best Distance: {best_distance} Km')
    return best_route
print('Le programme prend du temps\nAttender la resultat...')
chemin=genetic_algorithm(students, 3, 0.1)
m=folium.Map(location=[32.987676,-7.637354],zoom_start=15)
for i in chemin :
    folium.Marker(list(i),popup=students_dict[i]).add_to(m)

folium.PolyLine(chemin).add_to(m)
```
